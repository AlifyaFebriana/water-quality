```python
import cv2
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path = '/content/drive/MyDrive/dataset/DATASET/IMG20211012173020.jpg'
```

```python
img = cv2.imread(path)
print(img.shape)
```

```python
width, height = 224, 224
imgResize = cv2.resize(img, (width, height))
```

```python
cv2.imwrite ('/content/61.jpg', imgResize)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.imshow(img)
plt.title('Original Image')
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.imshow(imgResize)
plt.title('Resize Image')
plt.show()
```
